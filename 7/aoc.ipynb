```python
f = open("input.txt", "r")
input = list(f.read().split("\n"))
```

```python
from functools import cmp_to_key

hands= [(str_val.split(" ")[0], int(str_val.split(" ")[1])) for str_val in input]
normal_order = "AKQJT98765432"

def get_distribution(hand, order):
    nums = [0 for _ in order]
    for char in hand:
        nums[order.index(char)] += 1
    return nums

def count(l, i):
    return sum([1 if v == i else 0 for v in l])

def get_order(hand):
    dist = get_distribution(hand, normal_order)
    return 6 if 5 in dist else 5 if 4 in dist else 4 if (3 in dist) and (2 in dist) else 3 if 3 in dist else 2 if count(dist, 2) == 2 else 1 if 2 in dist else 0

def compare_hands(hand1, hand2):
    order1, order2 = get_order(hand1[0]), get_order(hand2[0])
    if order1 != order2:
        return order1 - order2
    
    for char1, char2 in zip(hand1[0], hand2[0]):
        i1, i2 = normal_order.index(char1), normal_order.index(char2)
        if i1 == i2:
            continue
        return 1 if i1 < i2 else -1

def GetSolution(hands, cmp):
    sorted_hands = sorted(hands, key = cmp_to_key(cmp))

    summation = 0
    for i, (hand, bet) in enumerate(sorted_hands):
        summation += (i + 1) * bet
    print(f"Solution: {summation:,}")
    return sorted_hands

sorted_cards = GetSolution(hands, compare_hands)
```

```python
joker_order = "AKQT98765432J"

def remap_jokers(hand):
    dist = get_distribution(hand, joker_order)
    max_val = max(dist[:-1]) # <-- This [:-1] is what I forgot and what doubled my time.
    max_index = dist.index(max_val)
    max_char = joker_order[max_index]
    return "".join([c if not c == "J" else max_char for c in hand])

def compare_hands_joker(hand1, hand2):
    order1, order2 = get_order(remap_jokers(hand1[0])), get_order(remap_jokers(hand2[0]))
    if order1 != order2:
        return order1 - order2
    
    for char1, char2 in zip(hand1[0], hand2[0]):
        i1, i2 = joker_order.index(char1), joker_order.index(char2)
        if i1 == i2:
            continue
        return 1 if i1 < i2 else -1




print([(hand, remap_jokers(hand)) for hand, _ in hands if "J" in hand])

print(GetSolution(hands, compare_hands_joker))


    
```
